```python
bike_share = pd.read_csv("Data/bikeshare.csv",header=0)
bike_share.head()
```

```python
# Aggregate on the day level
bike_day = bike_share.groupby(['dteday']).agg({'temp': 'mean', 'atemp': 'mean', 'hum': 'mean','windspeed': 'mean', 'casual': 'sum', 'registered': 'sum','cnt': 'sum',   })
```

#Temperature and Usage
Looking at the histograms above, usage appears to have some relationship to the actual tempature. There's also a difference between the two types of riders. A larger share of registered users take bikes out in colder temperatures, when compared to casual users.  This could be tied to external factors, such as an influx of tourists using bikes in warmer months, or could simply reflect different types of usage (e.g. exploration for casual users versus commuting for registered users).

```python
##Looking by month
print "Order of boxplots:"
for r in riders:
    sns.set_style('white') # sets seaborn "white" to the default matplotlib style
    plt.figure(figsize=(11, 6), dpi=80)
    plt.subplot(1, 1, 1)
    print '* %s' % r
    plt.hist(bike_share.mnth, bins=12, weights=bike_share[r])
```

#Months and Usage
Here a slight difference in shape between casual riders and registered users indicates how these two groups differ.  As previously mentioned, it appears as though casual riders are using the bikes in warmer months. This can be contrasted with registered users, where the frequency of use still dips in colder months, but not as drastically. This may relate to the number of registered users who use the bike system for their commute.

```python
print "Order of boxplots:"
for r in riders:
    sns.set_style('white') # sets seaborn "white" to the default matplotlib style
    plt.figure(figsize=(12, 6), dpi=80)
    plt.subplot(1, 1, 1)
    print '* %s' % r
    plt.hist(bike_share.weekday, bins=7, weights=bike_share[r])
```

#Casual Weekenders v. Commuters
This set of boxplots provides the clearest indication of a difference in bike usaged between registered and casual users.  Casual users in general use the bikes far more often on the weekend, while registered users tend to use the bikes more frequently during the week.  This further supports the hypothesis that registered users are using the bikes for their commute or other weekday activities, while casual users are biking for tourism or city exploration on the weekends.
