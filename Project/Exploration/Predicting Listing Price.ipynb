##Predicting AirBnB Price

I am interested in predicting the price (per night) of an AirBnB listing.  This could be useful for all three parties involved in an AirBnB transaction:

1. **The Lister** - a predictive model could help new lister determine a target price per night, by feeding their listing information into the model.
2. **AirBnb** - similarly, AirBnB, which makes it's money off a commision, could use this model to predict the expected value of a new listing.  Combined with expected number of bookings within a time frame, they could use a model predicting listing price to set expectations how much incremental revenue to expect.
3. **Tourists** - finally, a tourist might use this model in a somewhat counter-intuitive way.  Assuming we can produce a relatively strong model, tourists could use this to find value within existing listings. By comparing predicted and actual prices for existing listings, a tourist could look for the largest deltas in the data-set.  Listings at lower than predicted prices may represent under-priced listings, and prove the best bang for your buck.

```python
##Remove Cols with insufficient data or too many missing variablest
cols = [                          
'latitude',                       
'longitude',                      
'accommodates',                     
'bathrooms',                      
'bedrooms',                       
'beds',                                                                 
'guests_included',                  
'extra_people',                   
'minimum_nights',
'maximum_nights',
'availability_30',                  
'availability_60',                  
'availability_90',                  
'availability_365', 
'number_of_reviews', 
'host_listing_count',
'price',
'log_price'
] 

test = df[cols].dropna()

features = [                          
'latitude',                       
'longitude',                      
'accommodates',                     
'bathrooms',                      
'bedrooms',                       
'beds',                                                                 
'guests_included',                  
'extra_people',                   
'minimum_nights',
'maximum_nights',
'availability_30',                  
'availability_60',                  
'availability_90',                  
'availability_365', 
'number_of_reviews', 
'host_listing_count',
] 

x = test[features]     
y = test['log_price']
x.describe()
```

```python
##Quickly confirm that these seem to have low pvals in predicting price
from sklearn import feature_selection as f_select

def sig_features(x, y, threshold=0.05):
   """Given a column list, a feature set, a target, and a p value threshold, return all columns that show low p-values.
   x         : dataframe containing the feature set
   y         : target values for the model
   threshold : default is 0.5, threshold p value that each feature must fall below
   """
   significant_features = []
   for feature in x:
       pval = f_select.f_classif(x[feature],y)
       if pval[1][0] < threshold:
           significant_features.append(feature)
   print '%d Features are worth further consideration. (With P-Values below %f)' %(len(significant_features), threshold) 
   return significant_features

print 'Started with %d features.' % (len(features))        
sig_features(x,y,threshold=0.0001)
```

##Paring Down

After looking at data-types, a describe, feature selection and scatter plots, I can eliminate features based four criteria.

1. **Categorical Data**: to use linear regression and predict price, I need continuous value data -this eliminates any non integer/float columns.
2. **Missing Values**: as A rule of thumb, I removed features that had missing values for > 10% of the data set.
3. **P-Value**: Any features that did not meet the p-value threshold using a function I built using the feature selection module form Sci-Kit Learn
4. **Shape**: using the scatter matrix, the distribution of some data doesn't yeild a clear relationship to price.  In the interest of improving model accuracy I will not include these either.
