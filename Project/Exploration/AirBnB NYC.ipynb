```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

sns.set_style('white')

pd.set_option('display.max_colwidth',52)
```

```python
d_loc = '../../../Data/NYC/'
abnb_flist = pd.read_csv(d_loc + 'nyc_airbnb_listings_full.csv')
abnb_flist.head()
```

```python
#ratings out of 10 or 100 so 10 bins seems reasonable
n_bins = 10
fig, axes = plt.subplots(nrows=6, ncols=2,)

##can i use a for loop here?
abnb_flist['review_scores_rating'].plot(ax=axes[0, 0], kind='kde')
abnb_flist['review_scores_rating'].hist(ax=axes[0, 1], bins=n_bins)

abnb_flist['review_scores_accuracy'].plot(ax=axes[1, 0], kind='kde')
abnb_flist['review_scores_accuracy'].hist(ax=axes[1, 1], bins=n_bins)

abnb_flist['review_scores_cleanliness'].plot(ax=axes[2, 0], kind='kde')
abnb_flist['review_scores_cleanliness'].hist(ax=axes[2, 1], bins=n_bins)

abnb_flist['review_scores_checkin'].plot(ax=axes[3, 0], kind='kde')
abnb_flist['review_scores_checkin'].hist(ax=axes[3, 1], bins=n_bins)

abnb_flist['review_scores_location'].plot(ax=axes[4, 0], kind='kde')
abnb_flist['review_scores_location'].hist(ax=axes[4, 1], bins=n_bins)

abnb_flist['review_scores_value'].plot(ax=axes[5, 0], kind='kde')
abnb_flist['review_scores_value'].hist(ax=axes[5, 1], bins=n_bins)

print '-' * 13
print 'AIRBNB SCORES'
print '-' * 13
print fig
```

```python
sns.set_style('whitegrid')

# using plot and the '.' notation.
plt.figure(figsize=(15, 10), dpi=120)
plt.subplot(1, 1, 1)
plt.plot(abnb_reduced.longitude, abnb_reduced.latitude, '.', alpha = 0.2)
```

```python
#Zoom in a little to where most coords are located
plt.figure(figsize=(15, 12), dpi=120)
plt.subplot(1, 1, 1)

abnb_longlat1 = abnb_reduced[abnb_reduced['longitude'] > -74.05]
abnb_longlat2 = abnb_longlat1[abnb_longlat1['latitude'] > 40.60]
abnb_longlat3 = abnb_longlat2[abnb_longlat2['longitude'] < -73.85]
abnb_longlat = abnb_longlat3[abnb_longlat3['latitude'] < 40.90]

plt.plot(abnb_longlat.longitude, abnb_longlat.latitude, '.', alpha = 0.3, color = 'g')
```

```python
high_rate = abnb_longlat[abnb_longlat['review_scores_rating'] > 84]

low_rate = abnb_longlat[abnb_longlat['review_scores_rating'] < 70]

med_rate1 = abnb_longlat[abnb_longlat['review_scores_rating'] > 69]
med_rate = med_rate1[med_rate1['review_scores_rating'] < 85]

plt.rcParams['figure.figsize'] = (15, 12)                   # Set default figure size values for the entire notebook

fig, ax = plt.subplots(1)
alpha = 0.15

ax.scatter(high_rate.longitude, high_rate.latitude, label='High Rated', color='b', alpha=alpha)
ax.scatter(med_rate.longitude, med_rate.latitude, label='Medium Rated', color='m', alpha=alpha)
ax.scatter(low_rate.longitude, low_rate.latitude, label='Low Rated', color='r', alpha=alpha)

plt.title('Rating by Location', fontsize = 20)
plt.legend(["High Rated", "Medium Rated", "Low"])
```
