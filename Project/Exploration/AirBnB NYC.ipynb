```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

sns.set_style('white')
```

```python
d_loc = '../../Data/NYC/'
abnb_flist = pd.read_csv(d_loc + 'nyc_airbnb_listings_full.csv')
abnb_flist.head()
```

```python
abnb_flist.describe()
```

```python
rating_info = {
'review_scores_rating': 'OVERALL',
'review_scores_accuracy': 'ACCURACY',
'review_scores_cleanliness': 'CLEANLINESS',
'review_scores_checkin': 'CHECK-IN',
'review_scores_location': 'LOCATION',
'review_scores_value': 'VALUE',
}

for i in rating_info:
    print
    print '-' * 21
    print 'SCORE: %s' % rating_info[i]
    print '-' * 21
    print abnb_flist[i].describe()
```

```python
#ratings out of 10 or 100 so 10 bins seems reasonable
n_bins = 10
fig, axes = plt.subplots(nrows=6, ncols=2,)

##can i use a for loop here?
abnb_flist['review_scores_rating'].plot(ax=axes[0, 0], kind='kde')
abnb_flist['review_scores_rating'].hist(ax=axes[0, 1], bins=n_bins)

abnb_flist['review_scores_accuracy'].plot(ax=axes[1, 0], kind='kde')
abnb_flist['review_scores_accuracy'].hist(ax=axes[1, 1], bins=n_bins)

abnb_flist['review_scores_cleanliness'].plot(ax=axes[2, 0], kind='kde')
abnb_flist['review_scores_cleanliness'].hist(ax=axes[2, 1], bins=n_bins)

abnb_flist['review_scores_checkin'].plot(ax=axes[3, 0], kind='kde')
abnb_flist['review_scores_checkin'].hist(ax=axes[3, 1], bins=n_bins)

abnb_flist['review_scores_location'].plot(ax=axes[4, 0], kind='kde')
abnb_flist['review_scores_location'].hist(ax=axes[4, 1], bins=n_bins)

abnb_flist['review_scores_value'].plot(ax=axes[5, 0], kind='kde')
abnb_flist['review_scores_value'].hist(ax=axes[5, 1], bins=n_bins)

print '-' * 13
print 'AIRBNB SCORES'
print '-' * 13
print fig
```
