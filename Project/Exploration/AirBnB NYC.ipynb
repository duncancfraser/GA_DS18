#Some early observations
In starting to understand this data set, I though looking at the customer experience would be interesting.  I wanted to start to look at what factors might influence that experience. 

A quick explanation of the different scores:

* Rating (review_scores_rating) - this is the overall listing rating. It ranges from 0 to 100, with a skew toward higher scores.  I suspect AirBnB makes it hard to list if you continue to recieve poor ratings, which might explain why the mean is around 92.
* Accuracy (review_scores_accuracy) - this is a 10 point scale for accuracy of listing, that is did the lister accurately describe what the customer paid for?
* Cleanliness (review_scores_cleanliness) - pretty clear, 10 point rating on cleanliness of rating.
* Checkin (review_scores_checkin) - this is a 10 point scale for checkin experience.
* Location (review_scores_location) - this is a 10 point scale for location.
* Value (review_scores_value) - this is a 10 point scale for overall value.

Based on what we see from the scatter matrix, the various 10 point scales seem to all contribute to the overall rating. 
I was also curious how much space (which is at a premium in New York) influenced overall Rating. Based on the scatter plot above, I see some clustering of more spacious listings with higher ratings, but visually, it doesn't look quite as strong as I might expect.

```python
fig, ax = plt.subplots(1)

plt.subplot(1, 1, 1)
plt.plot(abnb_reduced.longitude, abnb_reduced.latitude, '.', alpha = 0.2)

plt.subplot(1, 1, 1)

plt.title('Scatter Map', fontsize = 24)

print fig
```

```python
#Zoom in a little to where most coords are located

long_min = -74.05
long_max = -73.85
lat_min = 40.60
lat_max = 40.90

abnb_longlat1 = abnb_reduced[abnb_reduced['longitude'] > long_min]
abnb_longlat2 = abnb_longlat1[abnb_longlat1['latitude'] > lat_min]
abnb_longlat3 = abnb_longlat2[abnb_longlat2['longitude'] < long_max]
abnb_longlat = abnb_longlat3[abnb_longlat3['latitude'] < lat_max]

fig, ax = plt.subplots(1)

plt.subplot(1, 1, 1)

plt.title('Hello, New York!', fontsize = 24)
plt.plot(abnb_longlat.longitude, abnb_longlat.latitude, '.', alpha = 0.3)

print fig
```

```python
upper_lim = 95
lower_lim = 85

high_rate = abnb_longlat[abnb_longlat['review_scores_rating'] >= upper_lim]
low_rate = abnb_longlat[abnb_longlat['review_scores_rating'] < lower_lim]
med_rate1 = abnb_longlat[abnb_longlat['review_scores_rating'] >= lower_lim]
med_rate = med_rate1[med_rate1['review_scores_rating'] < upper_lim]

fig, ax = plt.subplots(1)
alpha = 0.15

ax.scatter(high_rate.longitude, high_rate.latitude, label='High Rated', color='b', alpha=alpha)
ax.scatter(med_rate.longitude, med_rate.latitude, label='Medium Rated', color='g', alpha=alpha)
ax.scatter(low_rate.longitude, low_rate.latitude, label='Low Rated', color='r', alpha=alpha)

plt.title('Rating by Location', fontsize = 24)
plt.legend(["High Rated", "Medium Rated", "Low"])

print fig
```
