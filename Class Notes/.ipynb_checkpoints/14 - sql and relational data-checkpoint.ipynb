##9. Which school has generated the most distinct players?

```python
##SQL Style
query = 'SELECT schoolid,COUNT(distinct playerid) FROM collegeplaying GROUP BY schoolid ORDER BY COUNT(distinct playerid) DESC' 
college_players = pd.read_sql_query(query, cnx)
```

```python
college_players.head()
```

##10. Which school has generated the most expensive players? (expensive defined by their first year's salary).

```python
query  = "SELECT a.schoolid,AVG(b.salary) FROM collegeplaying a INNER JOIN (SELECT c.playerid,c.salary FROM salaries c WHERE c.yearid = (SELECT MIN(d.yearid) FROM salaries d where c.playerid=d.playerid)) b ON a.playerid = b.playerid GROUP BY a.schoolid ORDER BY avg(b.salary) DESC"
college_salaries = pd.read_sql_query(query, cnx)
college_salaries.head()
```

##11. Show the 5 most expensive salaries for each team in the year 2014.

##12. Partition the average salaries by team and year, against year. Find players that were paid more than 1 standard deviation above the average salary for that team and year. Show a count by playerid.

```python
query = 'SELECT distinct yearid,teamid,AVG(salary) over (PARTITION BY yearid, teamid),AVG(salary) over (PARTITION BY yearid) FROM salaries'
avg_tm_salary = pd.read_sql_query(query, cnx)
avg_tm_salary.head()
```
