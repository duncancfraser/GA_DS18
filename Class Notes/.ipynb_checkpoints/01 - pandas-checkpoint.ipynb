```python
header_row = ['age',   #naming a list for the header row
              'sex', 
              'cp', 
              'trestbps',
              'chol', 
              'fbs', 
              'restecg', 
              'thalach', 
              'exang', 
              'oldpeak', 
              'slope', 
              'ca', 
              'thal', 
              'num',]

url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data' #data set url
heart_data = pd.read_csv(url, header=0, na_values='?')
heart_data.columns = header_row #assigns the column names based on the header row named above
```

```python
""" Trial 1: Drop missing values. 
Pandas defines this on the axis:
1 = along the columns
0 = along the rows
"""
print heart_data[285:290]

dropped_rows_heart_data = heart_data.dropna()
dropped_ca_thal_heart_data = heart_data.dropna(1)

print dropped_rows_heart_data[285:290]
print dropped_ca_thal_heart_data[285:290]

""" Trial 2: Fill the data
.ffill()  = forward
.bfill()  = backward
.fillna() = set value
"""
print heart_data[285:290]['ca'].ffill() # forward fills the column
print heart_data[285:290]['ca'].bfill() # back fills the column
print heart_data[285:290]['ca'].fillna(0) # fill with 0s
print heart_data[285:290]['ca'].fillna(int(heart_data['ca'].mode())) # fill with the most common value
```

```python
def high_resting(row):
    if row['trestbps'] > row['thalach']:
        return 1
    else:
        return 0

# axis = 1 means what again? axis = 0 means you're working with a series not a frame
heart_data['high_resting'] = heart_data.apply(high_resting, axis=1) # .apply used to put your own function in there

# This would work the same way!
heart_data['high_resting'] = heart_data.apply(lambda row: 1 if row['trestbps'] > row['thalach'] else 0, axis=1)

# axis = 0 means what again?
heart_data['resting2x'] = heart_data['trestbps'].apply(lambda x: x*2)
```
